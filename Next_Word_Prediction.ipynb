#Creating Bigrams

In the sentence "I love natural language processing," the bigrams are:

```
words = [each_word.lower() for each_sentence in corpus for each_word in each_sentence]
```

```
words
```

```
bigram_list = list(bigrams(words))
```

```
print(bigram_list[:10])
```

#Creating a Conditional Frquency Distribution

A Conditional Frequency Distribution (CFD) in NLTK is a collection of frequency distributions, each associated with a different condition. It allows you to count frequencies of events, grouped by some condition.

For example, you can use a CFD to count the frequency of words (events) in different genres (conditions) of a text corpus.

```
cfd  = ConditionalFreqDist(bigram_list)
```

```
cfd['a']
```

```
cfd['house']
```

```
cfd['the']
# ('the', 'company') -> 3126 times in the entire bigrams list
# ('the', 'u') -> 2264 times in the entire bigrams list
# ('the', 'dollar') -> 984 times in the entire bigrams list
# ('the', 'bank') -> 960 times in the entire bigrams list
# ('the', 'first') -> 839 times in the entire bigrams list
```

#Predicting the Next Word

```
def Predict_next_word(input_word):
    input_word = input_word.lower()
    if input_word in cfd:
        return cfd[input_word].max()
    else:
        return 'Word not found in corpus(dataset)'

```

```

input_word = input("Enter the word whose next word You want to predict:")
next_word = Predict_next_word(input_word)
print(f"The next word after '{input_word}' could be: {next_word}")
```
